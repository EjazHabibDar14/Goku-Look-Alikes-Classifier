```python
from duckduckgo_search import ddg_images
from fastcore.all import *

def search_images(term, max_images=300):
    print(f"Searching for '{term}'")
    return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

```python
searches1 = ["Turles"]
searches2 = ['Bardock', 'Goku Jr.', 'Goku Black']
searches3 = ['Goku', 'Goten']
path = Path('Goku Look Alikes')
from time import sleep

for o in searches1:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f'{o} Dragon Ball Z photos'))
    download_images(dest, urls=search_images(f'{o}  Tree of Might photos'))
    download_images(dest, urls=search_images(f'{o}  Super Saiyan photos'))
for o in searches2:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f'{o} photos'))
    download_images(dest, urls=search_images(f'{o}  Dragon Ball photos'))
    download_images(dest, urls=search_images(f'{o}  Super Saiyan photos'))
for o in searches3:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f'{o} photos'))
    download_images(dest, urls=search_images(f'{o}  kid photos'))
    download_images(dest, urls=search_images(f'{o}  GT photos'))

    sleep(10)
    resize_images(path/o, max_size=400, dest=path/o)
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(path, bs=64)

dls.show_batch(max_n=20)
```

```python
from fastai.vision.all import *
import os

# Set the path to your image directory
path = Path('Goku Look Alikes')

# Define the categories
categories = ('Goku', 'Goku Black', 'Turles', 'Bardock', 'Goku Jr.', 'Goten')

# Function to count pictures in each category
def count_pictures(category):
    count = len(get_image_files(path/category))
    print(f"Category: {category}")
    print(count)
    print("---")

# Calculate counts for each category
for category in categories:
    count_pictures(category)
```

```python
learn = vision_learner(dls, resnet50, metrics=error_rate)
learn.fine_tune(20)
```

```python
is_this_Goku_Black,_,probs = learn.predict(PILImage.create('Goku Black.jpg'))
print(f"This is a: {is_this_Goku_Black}.")
print(f"Probability this is Goku Black: {probs[0]:.4f}")
```

```python
from fastai.vision.all import *
import gradio as gr

def is_black(x):
  return x[0].issuper()
```

```python
im= PILImage.create('Goten.jpg')
im = im.resize((192, 192), resample=Image.LANCZOS)
im.show()
```

```python
learn =load_learner('model.pkl')
```

```python
learn.predict(im)
```

```python
categories=('Bardock', 'Goku', 'Goku Black', 'Goku Jr.', 'Goten', 'Turles')
def classify_image(img):
  pred,idx,probs = learn.predict(img)
  return dict(zip(categories,map(float,probs)))
```

```python
classify_image(im)
```

```python
preds, targets = learn.get_preds(dl=dls.valid)

predicted_labels = preds.argmax(1)

accuracy = (predicted_labels == targets).float().mean()
print(f"Accuracy on the validation dataset: {accuracy.item() * 100:.2f}%")
```

```python
image=gr.inputs.Image(shape=(192,192))
label=gr.outputs.Label()
examples =['Bardock.jpg', 'Goku.jpg', 'Goku Black.jpg', 'Goku Jr..jpg', 'Goten.jpg', 'Turles.jpg']

intf = gr.Interface(fn=classify_image, inputs=image, outputs=label, examples=examples)
intf.launch(inline=False, share=True)
```
